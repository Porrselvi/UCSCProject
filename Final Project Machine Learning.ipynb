```python
from skimage.feature import greycomatrix, greycoprops
from skimage.color import rgb2gray

gray = (rgb2gray(img) * 255).astype(np.uint8)  # convert to 0-255 integer

# Compute GLCM
glcm = greycomatrix(gray, distances=[1], angles=[0], levels=256, symmetric=True, normed=True)

# Extract properties
contrast = greycoprops(glcm, 'contrast')[0,0]
dissimilarity = greycoprops(glcm, 'dissimilarity')[0,0]
homogeneity = greycoprops(glcm, 'homogeneity')[0,0]
energy = greycoprops(glcm, 'energy')[0,0]
correlation = greycoprops(glcm, 'correlation')[0,0]

print("Texture features:")
print("Contrast:", contrast)
print("Dissimilarity:", dissimilarity)
print("Homogeneity:", homogeneity)
print("Energy:", energy)
print("Correlation:", correlation)
```

```python
import os
import pandas as pd
from skimage import io, color
from skimage.transform import resize
import numpy as np
import matplotlib.pyplot as plt

# 1. Folder containing images
image_folder = r"C:\Users\APP\Pictures\Neem Leaf\New Img\Appam"
output_file = r"C:\Users\APP\Documents\Appam_pixel_values.xlsx"

# 2. Prepare a list to store flattened images
image_data = []

# 3. Loop through all images in folder
for filename in os.listdir(image_folder):
    if filename.endswith(".jpg") or filename.endswith(".png"):
        img_path = os.path.join(image_folder, filename)
        
        # Load image
        img = io.imread(img_path)
        
        # Resize image (optional, e.g., 128x128)
        img_resized = resize(img, (128, 128))
        
        # Convert to grayscale
        gray_img = color.rgb2gray(img_resized)
        
        # Flatten image into 1D array
        flat_img = gray_img.flatten()
        
        # Add to list
        image_data.append(flat_img)

# 4. Convert list to DataFrame
df = pd.DataFrame(image_data)

# 5. Save to Excel
df.to_excel(output_file, index=False)

print(f"Processed {len(image_data)} images and saved to {output_file}")

# 6. Optional: display first image
plt.imshow(image_data[0].reshape(128,128), cmap='gray')
plt.axis('off')
plt.title("First Processed Image")
plt.show()
```

```python
import os
import numpy as np
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Path to your image folders (each subfolder = class)
data_dir = r"C:/Users/APP/Pictures/Neem Leaf/New Img"

# Define image size
img_height, img_width = 128, 128
batch_size = 32

# Image augmentation and normalization
datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2,   # 20% for validation
    rotation_range=20,
    width_shift_range=0.1,
    height_shift_range=0.1,
    horizontal_flip=True
)

# Training data
train_data = datagen.flow_from_directory(
    data_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='categorical',
    subset='training'
)

# Validation data
val_data = datagen.flow_from_directory(
    data_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='categorical',
    subset='validation'
)
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models

num_classes = len(train_data.class_indices)

model = models.Sequential([
    layers.Conv2D(32, (3,3), activation='relu', input_shape=(img_height, img_width, 3)),
    layers.MaxPooling2D((2,2)),
    
    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),
    
    layers.Conv2D(128, (3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),
    
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(num_classes, activation='softmax')
])

model.summary()
```

```python
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```
